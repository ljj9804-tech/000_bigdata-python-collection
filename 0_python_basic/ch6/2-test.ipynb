```python
import pymysql
con = pymysql.connect(host='127.0.0.1', user='webuser',
                      password='webuser', database='webdb', charset='utf8')
print('DB 연결 성공!')
# con.close()
```

```python
cur = con.cursor()
cur.execute("""
CREATE TABLE product (
    id INT AUTO_INCREMENT PRIMARY KEY,
    name VARCHAR(100),
    price INT
)
""")
print('테이블 생성 완료')
```

```python
sql = "INSERT INTO product (name, price) VALUES (%s, %s)"
cur.execute(sql, ('노트북', 1200000))
con.commit()
print('삽입 완료')
```

```python
cur.execute('SELECT * FROM product WHERE price >= %s', (1000000,))
for row in cur.fetchall():
    print(row)
```
